```python
# 7
df['loan_status'].value_counts()
sns.countplot(x='loan_status' , data=df)
plt.show()
```

```python
# 8
Q8= df.groupby('Grade')['Intrest Rate'].mean().reset_index()
sns.barplot(x='Grade' , y='Intrest Rate' , data=Q8)
plt.xlabel('TYPES OF LOANS')
plt.ylabel('AVG INTREST RATE')
plt.show()
```

```python
# 9 
Q9= df.groupby('Emp_Lenght')['Intrest Rate'].mean().reset_index()
sns.lineplot(x='Emp_Lenght' , y='Intrest Rate' , data= Q9)
plt.xlabel('EXP OF EMPLOYEE')
plt.ylabel('AVG INTREST RATE')
plt.show()
```

```python
# 10
Q10= df.groupby('Annual Income')['Intrest Rate'].value_counts().reset_index()
sns.scatterplot(x='Intrest Rate' , y='Annual Income' , data= Q10)
plt.ylabel('Annual Income')
plt.xlabel('AVG INTREST RATE')
plt.show()
```

```python
# 11
Q11= df.groupby('loan_status')['Home Type'].value_counts().reset_index()
sns.lineplot(x='loan_status' , y='Home Type' , data= Q11)
plt.xlabel('STATUS OF LOAN')
plt.ylabel('TYPES OF HOMES')
plt.show()
```

```python
# 12
Q12= df.groupby('Grade')['Intrest Rate'].mean().reset_index()
sns.lineplot(x='Grade' , y='Intrest Rate' , data=Q12)
plt.xlabel('GRADES OF LOAN')
plt.ylabel('AVG INTREST LOAN')
plt.show()
```

```python
#13
Q13=df.groupby('Terms in Months')['Pending Payment'].mean().reset_index()
sns.barplot(x='Terms in Months' , y='Pending Payment' , data=Q13)
plt.xlabel('Terms in Months')
plt.ylabel('PENDING PAYMENTS')
plt.show()
```

```python
#14
Q14=df.groupby('DTI')['Loan Amount'].mean().head(20).reset_index()
sns.lineplot(x='DTI' , y= 'Loan Amount' , data=Q14)
plt.xlabel('DTI')
plt.ylabel('Laon Amount Avg')
plt.show()
```

```python
#15
Q15= df.groupby('Subgrade')['Intrest Rate'].mean().tail(10).reset_index()
sns.barplot(x='Subgrade' , y='Intrest Rate' , data=Q15)
plt.xlabel('TYPES OF LOANS (Subgrade)')
plt.ylabel('AVG INTREST RATE')
plt.show()
```

```python
# Q16
Q16=df[['Amount Taken', 'Loan Amount']].corr()
sns.scatterplot(x='Amount Taken' , y='Loan Amount' , data=Q16)
plt.xlabel('Amount Taken')
plt.ylabel('Loan Amount')
plt.show()
```

```python
#17
Q17= df.groupby('Home Type')['Loan Amount'].mean().reset_index()
sns.barplot(x='Home Type' , y='Loan Amount' , data=Q17)
plt.xlabel('TYPES OF LOANS')
plt.ylabel('AVG INTREST RATE')
plt.show()
```

```python
#18
Q18= df.groupby('verification_status')['Loan Amount'].sum().reset_index()
sns.lineplot(x='verification_status', y='Loan Amount', data=Q18)
plt.xlabel('Verification Status')
plt.ylabel('Total Loan Amount')
plt.show()
```

```python
#19
Q19= df.groupby('Subgrade')['Loan Amount'].sum().head(10).reset_index()
sns.barplot(x='Subgrade' , y='Loan Amount' , data=Q19)
plt.xlabel('TYPES OF LOANS (Subgrade)')
plt.ylabel('TOTAL LOAN AMOUNT')
plt.show()
```

```python
# 20
Q20 =df.groupby('DTI Check')['Loan Amount'].sum().reset_index()
sns.barplot(x='DTI Check' , y='Loan Amount' , data=Q20)
plt.xlabel('TYPES OF DTI (DTI Meter)')
plt.ylabel('Loan Amount')
plt.show()
```

```python
# 21
Q21= df.groupby('Annual Income')['Loan Amount'].sum().reset_index()
sns.scatterplot(x='Annual Income' , y='Loan Amount' , data=Q21)
plt.xlabel('Annual Income')
plt.ylabel('Loan Amount')
plt.show()
```

```python
# 22
Q22 = df.groupby('Purpose')['Loan Amount'].sum().reset_index()

plt.pie(Q22['Loan Amount'], labels=Q22['Purpose'], autopct='%1.1f%%', startangle=140)
plt.title('Total Loan Amount by Purpose')
plt.axis('equal')
plt.show()
```
